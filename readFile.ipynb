```python
import struct
from pathlib import Path

def read_raw_file(filepath, show_samples=5):
    """Читает .raw файл и выводит все поля"""
    path = Path(filepath)
    if not path.exists():
        raise FileNotFoundError(f"❌ Файл не найден: {path}")

    with open(path, 'rb') as f:
        data = f.read()

    if len(data) < 17:
        raise ValueError("⚠️ Файл слишком короткий. Ожидается минимум 17 байт заголовка.")

    # Java DataOutputStream использует Big-Endian (>)
    # i = int (4 байта), d = double (8 байт), B = unsigned byte (1 байт)
    x = struct.unpack_from('>i', data, 0)[0]
    y = struct.unpack_from('>i', data, 4)[0]
    thicknom = struct.unpack_from('>d', data, 8)[0]
    defective = struct.unpack_from('>B', data, 16)[0]

    num_values = x * y
    expected_size = 17 + num_values * 8
    
    # Безопасное извлечение матрицы
    if len(data) < expected_size:
        print(f"⚠️ Файл обрезан. Ожидается {expected_size} байт, прочитано {len(data)}.")
        available = (len(data) - 17) // 8
        matrix = struct.unpack(f'>{available}d', data[17:])
    else:
        matrix = struct.unpack(f'>{num_values}d', data[17:17+num_values*8])

    # Вывод информации
    print("="*60)
    print("📊 ИНФОРМАЦИЯ О .raw ФАЙЛЕ")
    print("="*60)
    print(f"📐 Размеры матрицы (X × Y) : {x} × {y}")
    print(f"🔢 Всего элементов          : {len(matrix)}")
    print(f"🔩 Номинальная толщина       : {thicknom:.4f}")
    print(f"🚩 Признак брака (defective) : {defective} ({'БРАК' if defective == 1 else 'ГОДНЫЙ'})")
    print("-"*60)

    if len(matrix) > 0:
        print(f"📈 Минимум                 : {min(matrix):.4f}")
        print(f"📉 Максимум                : {max(matrix):.4f}")
        print(f"📊 Среднее                 : {sum(matrix)/len(matrix):.4f}")
        
        if show_samples > 0:
            print(f"👀 Первые {show_samples} значений : {matrix[:show_samples]}")
            if len(matrix) > show_samples * 2:
                print(f"👀 Последние {show_samples} значений: {matrix[-show_samples:]}")
    print("="*60)
    
    return {
        'x': x, 'y': y, 'thicknom': thicknom, 
        'defective': defective, 'data': matrix
    }

# 🚀 ЗАПУСК (укажите путь к вашему файлу)
read_raw_file('good_samples/thick_378463.raw', show_samples=25)
```
