```python
from pathlib import Path

def patch_existing_folders(good_dir='good_samples', bad_dir='bad_samples'):
    good_path = Path(good_dir)
    bad_path = Path(bad_dir)
    
    processed = 0
    skipped = 0
    
    for folder, is_defective in [(good_path, False), (bad_path, True)]:
        if not folder.exists(): continue
        print(f"\n📂 Обработка: {folder}")
        
        for raw_file in folder.glob('*.raw'):
            try:
                with open(raw_file, 'rb') as f:
                    data = bytearray(f.read())
                    
                if len(data) < 16: continue
                
                # Проверка на повторный патчинг
                if (len(data) - 17) % 8 == 0:
                    skipped += 1
                    continue
                    
                data.insert(16, 1 if is_defective else 0)
                
                with open(raw_file, 'wb') as f:
                    f.write(data)
                processed += 1
            except Exception as e:
                print(f"   ⚠️ Ошибка {raw_file.name}: {e}")
                
    print(f"\n✅ Готово! Обработано: {processed}, Пропущено (уже пропатчены): {skipped}")

# Запуск патчера для уже готовых папок
patch_existing_folders()
```
